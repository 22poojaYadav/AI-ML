```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x=np.array([[95,90],[85,82],[60,55],[75,70],[65,65],[81,56],[99,89],[47,55],[30,33],[40,50],[72,62],[69,89],[79,36],[62,78],[55,66]])
```

```python
km=KMeans(n_clusters=3,random_state=42)
km.fit(x)
```

```python
xyz=km.labels_ #each clusters beongs to whichh data
abc=km.cluster_centers_ #means of the all data
```

```python
#plot graph 
plt.figure(figsize=(8,8))
plt.scatter(x[:,0],x[:,1],c=xyz,cmap='Set1',)
plt.scatter(abc[:,0],abc[:,0],c='blue',marker='o')
plt.title('students Report')
plt.xlabel('Maths Marks')
plt.ylabel('science Marks')
plt.show()
```

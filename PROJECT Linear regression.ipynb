```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([[1],[2],[3],[4],[5]])
y=np.array([2,3,4,5,6])
x
```

```python
y
```

```python

xyz=LinearRegression()
xyz
```

```python
xyz.fit(x,y)
```

```python
pre=xyz.predict([[9]])
print(pre[0])
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,xyz.predict(x),color='blue')
plt.title('scatter plot for hours vs marks')
plt.xlabel('hours')
plt.ylabel('Marks')
plt.show()
```

```python
xyz.intercept_
```

```python
xyz.coef_
```

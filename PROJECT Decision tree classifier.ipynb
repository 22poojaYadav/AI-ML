```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score 
```

```python
df=pd.read_csv('salaries.csv')
df
```

```python
df['company']=df['company'].map({'google':0,'facebook':1,'abc pharma':1})
df['job']=df['job'].map({'sales executive':0,'computer programmer':1,'business manager':1})
df['degree']=df['degree'].map({'masters':0,'bachelors':0})
```

```python
x=df.drop(columns='salary_more_then_100k')#features of x
y=df['salary_more_then_100k'] #target
```

```python
#70% data for training &30% data for testing
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# creating model and traing 
abc=DecisionTreeClassifier(max_depth=2,random_state=42)
abc.fit(x_train,y_train)
```

```python
prediction=abc.predict(x_test)
prediction
```

```python
#evaluation
accuracy=accuracy_score(y_test,prediction)
precision=precision_score(y_test,prediction)
recall=recall_score(y_test,prediction)
f1=f1_score(y_test,prediction)
```

```python
# visualize the tree
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(15,9))
plot_tree(abc,feature_names=x.columns,class_names=['bachalorss','masters'],filled=True)
plt.title("Decision Tree for salaries")
plt.show()
```
